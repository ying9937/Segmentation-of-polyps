```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import os
import numpy as np
import h5py 
```

```python
# 读取路径，获得图片名称
file_path = os.path.join(os.getcwd(),r'F:\wujieyu\img test\TestDataset\Kvasir\images\img')
file_names = os.listdir(file_path)
```

```python
# 得到每张图片的路径
img_paths = []
for file_name in file_names:
    img_paths.append(os.path.join(file_path,file_name))
len(img_paths)

for i in img_paths:
    print(i)
```

```python
# 彩图
# 读取每一张图片
img1 = np.zeros((len(file_names),288, 384, 3))
for i in range(len(file_names)):
    img = cv2.imread(img_paths[i])
    img = cv2.resize(img,(384,288))
    img1[i] = img/255
```

```python
# 读取处理后的边缘图片
# 读取路径，获得图片名称
file_path = os.path.join(os.getcwd(),r'F:\wujieyu\img test\TestDataset\Kvasir\img edge\Sobel\bianyuan')
file_names = os.listdir(file_path)
# 得到每张图片的路径
Roberts_paths = []
for file_name in file_names:
    Roberts_paths.append(os.path.join(file_path,file_name))
len(Roberts_paths)

for i in Roberts_paths:
    print(i)
```

```python
# 读取每一张边缘图片
Roberts1 = np.zeros((len(file_names),288, 384, 3))
for i in range(len(Roberts_paths)):
    Roberts = cv2.imread(Roberts_paths[i])
    Roberts = cv2.resize(Roberts,(384,288))
    Roberts1[i] = Roberts/255
```

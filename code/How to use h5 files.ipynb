```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import os
import numpy as np
```

```python
file_path = os.path.join(os.getcwd(),r'C:\Users\lenovo\Desktop\Sample dataset')
file_names = os.listdir(file_path)
```

```python
img_paths = []
for file_name in file_names:
    img_paths.append(os.path.join(file_path,file_name))
len(img_paths)
```

```python
img_png =  np.random.rand(len(file_names), 288, 384, 3).astype('float32')
img_png.shape
```

```python
for i in range(len(img_paths)):
    img = cv2.imread(img_paths[i])
    img_re = cv2.resize(img,(384,288))
    img1 = img_re[...,::-1]
    img_png[i] = img1/255
```

```python
plt.imshow(img_png[0])
```

```python
# h5
import time
from tqdm import trange
import h5py
for i in trange(10):
    f = h5py.File(r'C:\Users\lenovo\Desktop\Sample-dataset.h5', 'w')
    f.create_dataset('img', data=img_png)
    f.close()
    time.sleep(0.5)
print("over")
```

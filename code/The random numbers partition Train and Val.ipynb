```python
import numpy as np
# 生成原始范围数组（0到2986）
low, high = 0, 3540
original = np.arange(low, high)
```

```python
# 生成不重复的随机数组并排序（示例）
sample_size = 540
np_random_unique = np.random.choice(original, size=sample_size, replace=False)
sorted_array = np.sort(np_random_unique)
```

```python
sorted_array.shape
```

```python
sorted_array
```

```python
# 找出原始范围中存在但排序数组中缺失的数字，并反向排列
missing_numbers = np.setdiff1d(original, sorted_array)  # 差集（默认升序）
# reverse_missing = missing_numbers[::-1]  # 反转实现降序
```
